Using Monte Carlo simulation, find the value of PI and area of the circle using the given circle and square. You have to simulate the value for n=100,1000,5000 and 10000 trials. Show the scatter plot, value of PI, value of the area for each value of n.
At the end of the simulation, draw two-bar diagrams. 

First Bar Plot:  x-axis: number of trials, y-axis: PI -value

Second Bar Plot:  x-axis: number of trials, y-axis: Area of the circle



```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
trial_list= [100, 1000, 5000, 10000]
pi_list = []
area_list = []

for n in trial_list:
    print("\n---------------------------------------------------------\n")
    hits=0
    x_hits = []
    y_hits = []
    x_miss = []
    y_miss = []
    for i in range(n):
        x= np.random.uniform(-5, 5)
        y= np.random.uniform(-5, 5)

        if math.sqrt(x**2 + y**2)<=3:
            hits= hits+1
            x_hits.append(x)
            y_hits.append(y)
        else:
            x_miss.append(x)
            y_miss.append(y)
        
    print("For Trials : ", n)
    PI= ((10**2)/(3**2)) * (hits/n)
    print("pi : ", PI)
    area_of_circle= (10**2) * (hits/n)
    print("Area of circle : ", area_of_circle)
    plt.scatter(x_hits, y_hits, color = "red", label = "Hit points")
    plt.scatter(x_miss, y_miss, color = "green", label = "Miss points")
    plt.legend()
    plt.show()
    #save the value in list
    pi_list.append(PI)
    area_list.append(area_of_circle)

print("\n----------------------------First bar plot-----------------------------\n")
trials = list(map(str, trial_list))
plt.bar(trials, pi_list)
plt.xlabel("Number of trials")
plt.ylabel("PI-value")
# plt.rcParams['figure.figsize'] = (10,10)
plt.show()

print("\n----------------------------Second bar plot-----------------------------\n")
trials = list(map(str, trial_list))
plt.bar(trials, area_list)
plt.xlabel("Number of trials")
plt.ylabel("area of circle")
# plt.rcParams['figure.figsize'] = (10,10)
plt.show()

```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
trial_list= [100,1000,5000,10000]

for n in trial_list:
    print("\n---------------------------------------------------------\n")
    hits=0
    x_hits = []
    y_hits = []
    x_miss = []
    y_miss = []
    for i in range(n):
        x= np.random.uniform(0,3)
        y= np.random.uniform(0,5)

        if y<= x+2:
            hits= hits+1
            x_hits.append(x)
            y_hits.append(y)
        else:
            x_miss.append(x)
            y_miss.append(y)
    print("For Trials : ", n)
    area_of_triangle= (3*5) * (hits/n)
    print("Area of Triangle : ", area_of_triangle)

    plt.scatter(x_hits, y_hits, color = "red", label = "Hit points")
    plt.scatter(x_miss, y_miss, color = "green", label = "Miss points")
    plt.legend()
    plt.show()
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    if(x<=4):
        return (4*x)**0.5   #y^2=4x  
    else:
        return (8-x)        #y = 8 - x

trial_list= [100,1000,5000,10000]

for n in trial_list:
    print("\n---------------------------------------------------------\n")
    hits=0
    x_hits = []
    y_hits = []
    x_miss = []
    y_miss = []
    for i in range(n):
        x= np.random.uniform(0,8)
        y= np.random.uniform(0,4)

        if y<= f(x):
            hits= hits+1
            x_hits.append(x)
            y_hits.append(y)
        else:
            x_miss.append(x)
            y_miss.append(y)
    print("For Trials : ", n)
    area_under_curve= (8*4) * (hits/n)
    print(hits)
    print("Area under curve : ", area_under_curve)

    plt.scatter(x_hits, y_hits, color = "red", label = "Hit points")
    plt.scatter(x_miss, y_miss, color = "green", label = "Miss points")
    plt.legend()
    plt.show()
```
